import numpy and matplotlib

```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt 
```

now define array 

```python
x = np.linspace(0,2*np.pi, 100)
print(x[-1],2*np.pi)

y = np.sin(x)
z = np.cos(x)
w = np.sin(4*x)
v = np.cos(4*x)
```

now lets make 2 panel plot side-by-side

```python
#call subplots to generate a multipanel figure. this means 1 row, 2 columns of figures
f, axarr = plt.subplots(1,2)

#treat axarr as an array, from left to right

#first panel 
axarr[0].plot(x,y)
axarr[0].set_label('x')
axarr[0].set_ylabel('sine(x)')
axarr[0].set_title(r'$\sin(x)$')
#second panel 
axarr[1].plot(x,z)
axarr[1].set_xlabel
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')
```

now we can adjust our sub plots

```python
#call subplots to generate a multipanel figure 
f. axarr = plt.subplots(1,2)

#treat axarr as an array, from left to right

#first panel 
axarr[0].plot(x,y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')
#second panel
axarr[1].plot(x,y)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

#add more space between the figures
f.subplots_adjust(wspace=0.4)
```

clearly these graphs have no graph, thats bc the info wasnt added to make the graph ut compare to earlier graph to see how to change the spacing between the graphs

lets now change the axis ratio and see what cool shit will happen

```python
#call subplots to generate a multipanel figure
f, axarr = plt.subplots(1, 2)
#treat axarr as an array

#first panel
axarr[0].plot(x,y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('cos(x)')
axarr[0].set_title(r'$\sin(x)$')

#second panel
axarr[1].plot(x,z)
axarr[1].set_label('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

f.subplots_adjust(wspace=0.4)

#fix the axis ratio
#here are two possible options
axarr[0].set_aspect('equal')#make the ratio of the tick units eual 
axarr[1].set_aspect(np.pi)#make a square by setting the aspect to be the ratio of the tick unit range
```
