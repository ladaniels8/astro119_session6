# exmpleof performing linear least squares fitting
first we import numpy and matplotlib as usual

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

noe lets generate some random data about a trend line

```python
#set a random number seed
np.random.seed(119)

#set number of data point
npoints = 50 

#set x 
x = np.linspace(0,10.,npoints)

#set slope, intercept, and scatter rms
m = 2.0 
b = 1.0 
sigma = 2.0 

#generate y points 
y = m*x + b +np.random.normal(scale=sigma,size=npoints)

#generate y points
y = m*x +b +np.random.normal(scale=sigma,size=npoints)
y_err = np.full(npoints,sigma)
```

```python
f = plt.figure(figsize=(7,7))
plt.errorbar(x,y,sigma,fmt='o')
plt.xlabel('x')
plt.ylabel('y')
```

method #1 polyfit()

```python
m_fit, b_fit = np.poly1d(np.polyfit(x, y, 1, w=1./y_err)) #weight with uncertainty, (alsot poly1d is poly one d, not an "L")
print(m_fit, b_fit)

y_fit = m_fit *x + b_fit
```

```python
f = plt.figure(figsize=(7,7))
plt.errorbar(x,y,yerr=y_err,fmt='o',label='data')
plt.plot(x,y_fit,label='fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2,frameon=False)
```
